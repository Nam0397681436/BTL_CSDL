```python
import pandas as pd
import numpy as np
import json


df_type=pd.read_csv(r"E:\PTIT\python\rawl_tiki\type_device_mobile.csv")
list_type=df_type.type.to_list()
for i in range(len(list_type)):
    list_type[i]=list_type[i].lower()
text=[
    ',',
    'ram',
    'rom',
    '|',
    '-'
]
def ChuanHoaPrice(s):

    s=s[:-1]
    s=s.replace('.','')
    return int(s)

def ChuanHoaQuantity(s):

    s = str(s)  
    if s != '0':  
        x = s.split(" ")  
        return x[0]  
    return s

def ChuanHoaEvaluate(s):
    s=str(s)
    if s!='0':
        return s[1:len(s)-1]
    return s  
  
def ChuanHoaName(s):
      tm=s.lower()
      for i in list_type:
          if i in tm:
            index=tm.find(i)
            return s[index:]
      return s  
    
def ChuanHoaName2(s):
    tm=s.lower()
    for i in text:
        if i in tm:
            index=tm.find(i)
            s=s[:index]
    tm1='điện thoại'
    return s

for i in range(1,14):
     
     url=r"E:\PTIT\python\lazada\page{}_lazada.csv".format(i)

     df=pd.read_csv(url)
     
     df['price'] = df['price'].apply(ChuanHoaPrice)
     df['quantity']=df['quantity'].apply(ChuanHoaQuantity).astype(int)
     df['evaluate']=df['evaluate'].apply(ChuanHoaEvaluate).astype(int)
     df['name']=df['name'].apply(ChuanHoaName)
     df['name']=df['name'].apply(ChuanHoaName2)

     df.to_csv('mobile_page{}_lazada.csv'.format(i),index=False)
    
     
```
