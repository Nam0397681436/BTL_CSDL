```python
import numpy as np
from selenium import webdriver
from time import sleep
import random
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
from selenium.webdriver.common.by import By

driver=webdriver.Chrome()



url='https://www.lazada.vn/dien-thoai-di-dong/?page=3&spm=a2o4n.searchList.cate_1.1.2608KxKZKxKZTE'

driver.get(url)

sleep(random.randint(5,10))

elements=driver.find_elements(By.CSS_SELECTOR, ".RfADt [href]") 

link= [element.get_attribute('href')   for element in elements]

```

```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

title = [element.text for element in elements]

def handle_discount(s):
    try:
        a = s.split(' ')

        if len(a) > 3:

            return a[2]
    except:

        return None

price_elements = driver.find_elements(By.CSS_SELECTOR, ".aBrP0")
price = [element.text for element in price_elements]

link = [element.get_attribute("href") for element in elements] 

df = pd.DataFrame(list(zip(title, link, price)), columns=['name', 'link', 'price'])


discounts_element = driver.find_elements(By.CSS_SELECTOR, '.WNoq3')
discount = [handle_discount(element.text) for element in discounts_element]

quantity_element = driver.find_elements(By.CSS_SELECTOR, '._6uN7R')

quantity = []
evaluate = []
address = []

for i in range(1,40 + 1):
    try:
        quantity.append(driver.find_element("xpath", "/html/body/div[4]/div/div[3]/div[1]/div/div[1]/div[2]/div[{}]/div/div/div[2]/div[5]/span[1]/span[1]".format(i)).text)
    except Exception as e:
        quantity.append("0")

for i in range(1,40 + 1):
    try:
        evaluate.append(driver.find_element("xpath","/html/body/div[4]/div/div[3]/div[1]/div/div[1]/div[2]/div[{}]/div/div/div[2]/div[5]/div/span".format(i)).text)
    except Exception as e:
        evaluate.append("0")

for i in range(1,40 + 1):
    try:
        address.append(driver.find_element("xpath", "/html/body/div[4]/div/div[3]/div[1]/div/div[1]/div[2]/div[{}]/div/div/div[2]/div[5]/span[2]".format(i)).text)
    except Exception as e:
        address.append(None)

df['quantity']=quantity
df['evaluate']=evaluate
df['address']=address

#print(quantity)
#print(evaluate)
#print(address)

df.index=df.index+1

df.to_csv('page1_lazada.csv',index=True)
```
